```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xi = 0.02
omega = 20
t = np.linspace(0, 10, 1000)
```

```python
free = np.exp(-xi*omega*t)*np.sin(omega*t)

plt.figure(facecolor='navajowhite', figsize=(20,1))
plt.margins(0,0)
plt.axis('off')
plt.plot(t, free, color='darkorange')
plt.savefig('free.png')
plt.show()
```

```python
force = np.sin(12*t) + 0.3*np.sin(3*t)
response = np.convolve(force, free, 'full') * 0.2
response = response[:len(response)//2+1]

plt.figure(facecolor='honeydew', figsize=(20,1))
plt.margins(0,0)
plt.axis('off')
plt.plot(t, force, color='darkgreen')
plt.plot(t, response, color='seagreen')
plt.savefig('forced.png')
plt.show()
```

```python
signal = 0.1*np.sin(9*t) + 0.2*np.sin(57*t) + 0.6*np.sin(83*t) + 0.3*np.sin(102*t)

plt.figure(facecolor='mistyrose', figsize=(20,1))
plt.margins(0,0)
plt.axis('off')
plt.plot(t, signal, color='maroon')
plt.savefig('mdof.png')
plt.show()
```
